```python
# option 1 - requires_grad_(False)
x=torch.tensor([1.0,2.0,3.0],requires_grad=True)
x
```

```python
y=x**2
y
```

```python
#Gradients are defined for scalar-valued functions
z=torch.sin(y).mean()
z
```

```python
# option 2 - detach()
x=torch.tensor([1.0,2.0,3.0],requires_grad=True)
y=x**2
z=torch.sin(y).mean()
z.backward()
x.grad
```

```python
k=x.detach()
```

```python
x.grad
```

```python
y1=k**2
y1
```

```python
y1.backward()
```

```python
# option 3 - torch.no_grad()

x=torch.tensor([1.0,2.0,3.0],requires_grad=True)
y=x**2
z=torch.sin(y).mean()
z.backward()
x.grad
```

```python
with torch.no_grad():
  y=x**2

y
```
