#Mathematical Operation

## Scalar Operation

```python
x=torch.rand(2,3)
x
```

```python
# addition
x + 2
# substraction
x - 2
# # multiplication
x * 3
# # division
x / 3
# # int division
(x*100)//3
# # mod
((x*100)//3)%2
# # power
x**2
```

## Element wise operation

```python
a=torch.rand(2,3)
a
```

```python
b=torch.rand(2,3)
b
```

```python
# add
a+b
# sub
a-b
# # multiply
a*b
# # division
a/b
# # power
a**b
# # mod
a%b
```

```python
c=torch.tensor([1,-2,3,-4])
d=torch.rand(2,3)+1
d
```

```python
# abs
torch.abs(c)

# negative
torch.neg(c)

# round
torch.round(d)

# ceil
torch.ceil(d)

# floor
torch.floor(d)

# clamp
torch.clamp(d,min=1.5,max=1.6)
```

## Reduction operation

```python
e=torch.randint(size=(2,3),low=2,high=8,dtype=torch.float16)
e
```

```python
# sum
torch.sum(e)

# sum along columns
torch.sum(e,dim=0)

# sum along rows
torch.sum(e,dim=1)

# mean
torch.mean(e)

# mean along col
torch.mean(e,dim=0)

# median
torch.median(e)

# max and min
torch.max(e,dim=0)
torch.min(e,dim=1)

# product
torch.prod(e,dim=1)

# standard deviation
torch.std(e,dim=0)

# variance
torch.var(e,dim=1)

# argmax
torch.argmax(e,dim=0)

# argmin
torch.argmin(e,dim=1)
```

## Comparison operations

```python
i = torch.randint(size=(2,3), low=0, high=10)
j = torch.randint(size=(2,3), low=0, high=10)

print(i)
print(j)
```

```python
# greater than
i > j
# less than
i < j
# equal to
i == j
# not equal to
i != j
# greater than equal to
i>=j
# less than equal to
i<=j
```

## Special functions

```python
a=torch.randint(size=(2,3),low=3,high=15,dtype=torch.float64)
a
```

```python
#log
torch.log(a)

#exp
torch.exp(a)

#sqrt
torch.sqrt(a)

#sigmoid
torch.sigmoid(a)

#softmax
torch.softmax(a,dim=0)

#relu
torch.relu(a)
```
